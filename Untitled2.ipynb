```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df=pd.read_csv('/content/in-vehicle-coupon-recommendation.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python

z = np.abs(stats.zscore(df.select_dtypes(include=[np.number])))


outliers = (z > 3).sum(axis=0)
outliers
```

```python
sns.boxplot(data=df.select_dtypes(include=[np.number]))
plt.title("Box Plot for Numerical Features")
plt.show()
```

```python
sns.scatterplot(data=df.select_dtypes(include=[np.number]))
plt.title("Scatter Plot for Numerical Features")
plt.show()
```
